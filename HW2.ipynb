```python
lists = []
x = int(input("자연수 한개를 입력하세요. "))
lists.append(x)
while(1):
  x = int(input("계속 입력을 원하시면, 자연수 한개를 입력하시고, 아니라면 0을 입력하세요. "))
  if x == 0:
    break
  lists.append(x)
  if len(lists) == 100:
    print("이제 그만하세요. 벌써 100개를 입력했어요.")
    break

temp = 0
for i in range(len(lists)-1):
  for j in range(i+1,len(lists)):
    if lists[i] < lists[j]:
      temp = lists[j]
      lists[j] = lists[i]
      lists[i] = temp

print("당신이 입력한 숫자는 큰 것부터 순서대로 말한다면 아래와 같습니다.")
for i in lists:
  print(i, end=" ")
  
```
