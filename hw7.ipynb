```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

targetUrl = 'https://raw.githubusercontent.com/Jung1016/python/main/caffeine.csv'

Data_set = pd.read_csv(targetUrl, sep = ',')

x = Data_set.iloc[:, 1:4]
y = Data_set.iloc[:, 4]

x_n =  (x - x.mean())/x.std()
y_d = pd.get_dummies(y)

model = Sequential()
model.add(Dense(25, input_dim=3, activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(6, activation = 'sigmoid'))
model.summary()

model.compile(loss = 'mse', optimizer = 'adam', metrics=['accuracy'])
history = model.fit(x_n, y_d, epochs = 150, batch_size = 5)

loss = model.evaluate(x_n,y_d)
y_predict = model.predict(x_n)
print(y_predict)

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```
